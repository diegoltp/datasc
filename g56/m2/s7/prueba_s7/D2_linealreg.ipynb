#### Definición del problema.
* Se busca generar un modelo que identifique los alumnos que presentan un bajo nivel academico.

#### Metodología
* Es un problema de regresión lineal con vector objetivo continuo.
* Estudiar los datos contenidos, densidades, segregaciones, agregaciones, sesgos mestadisticos descriptivos, estadisticos de dispersion, con respecto a variables y vector objetivo (income). 
* Describir la base de datos y re-definirla para adecuar sus valores, columnas, diseños, caracteres, etc...
* *Se analizan las distintas variables que pueden afectar el rendiemiento de cada estudiante.*
* *Por medio del análisis de vectores objetivos y variables correlacionadas al vector. Para ello primero se depura la base de datos entregada, se cumplira con la recodificacion solicitada además de los datos perdidos. Se adecúan los formatos que deban ser modificados para no tener problemas de codificación.*
* *Para las variables categoricas que se deban recodificar en binarias se aplica la metodologia Binary Encoding, con el fin de generar vectores perpendiculares entre si, sin colinealidad, es decir se genera  una matriz fullrank.*
* *Se visualizan las variables, medidas de tendencia central, de dispersion, densidad, frecuencia, correlacion, graficos de barras, histogramas, boxplot, entre otros para definir el comportamiento de vectores objetivo y variables.*
* *Mediante inferencia estadistica, pruebas de esfericidad, muestreo KMO, se evalua la posibilidad de extraer factores y reducir dimensionalidad, solo si el puntaje KMO es relevante en el ejercicio.*
* *Se analizará la correlacion entre variables y se seleccionaran las variables que continuan en analisis vs las que no.*
* *Se generan modelos depurados por medio de las metricas principales (BIC, Rsqr, P-value) para estudiar el comportamiento probabilistico de manera general.*
* *De ser significativos el objetivo sera estudiar un modelo depurado por p-value, 1 modelo a partir de las mejores correlaciones, y un modelo a partir del analisis de variables.*
* *Finalmente se crean y entrenan 3 modelos previamente definidos, para comparar sus metricas y evaluar su desempeño predictivo frente al vector objetivo promedio final. Con el fin de obtener el mejor modelo capaz de identificar si un estudiante presenta rendimiento deficiente.*
#### Métricas
* Metricas dimensionalidad:
    - Esfericidad de Bartlett: Identifica si la matriz de correlaciones es identidad o no, es decir las variables no estan relacionadas entre si. Se busca que la matriz no sea una matriz identidad.
    - Prueba de muestreo Kaiser-Meyer-Olkin (KMO): Eliminar elementos de la matriz que sean inestables y dificulten identificación de factores a extraer.
* Metricas Descriptivas: 
    - Indicadores BIC AUC: para medir la cantidad de parametros utilizados en los modelos. Mientras menor sea mejor.
    - P-Value: que nos indica si se puede rechazar la hipotesis nula. Ho indica que no se tienen valores significativos, por ende si se rechaza se tiene significancia en la variable o modelo de estudio.
    - Log-verosimilitud: Nos indica que tan bueno es el modelo en base a su verosimilitud, mientras mayor sea el indicador mejor.
    - R-squared: Capacidad explicativa del modelo y/o conjunto de variables.
* Metricas Prediccion:
    - R-Squared: Resume cuál es la capacidad explicativa de los regresores (variables independientes) en la variabilidad vel vector objetivo (variable dependiente).
    - neg_mean_squared_error (MSE): Es una métrica de error que mide la diferencia entre los valores reales y los valores previstos por el modelo. El valor negativo es una convención común y se utiliza para que un valor más bajo sea un mejor rendimiento.
    - explained_variance: Es un indicador que mide la variabilidad de los datos que es explicada por el modelo. Valores cercanos a 1 indican una buena capacidad del modelo para explicar los datos, mientras que valores cercanos a 0 indican una explicación pobre.
