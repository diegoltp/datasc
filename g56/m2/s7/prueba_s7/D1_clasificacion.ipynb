#### Definición del problema.
* Tarea encomendada:
    - Desarrollar un modelo predictivo sobre la probabilidad que un individuo presente salarios por sobre o bajo los 50.000 dólares anuales.
* Abordaremos el caso siguiendo la taxonomía de Hastie, Tibshinari y Friedman (2009), para realizar un modelo de aprendizaje donde el vector objetivo responde a un atributo discreto, es decir  evaluaremos la presencia o ausencia de un atributo, en este caso se evalúa si el individuo esta por sobre el margen de los 50000 o por de bajo de los 50000 usd anuales.   
#### Metodología
* Corresponde a un problema de clasificacion, donde el vector objetivo es una variable binaria.
* Estudiar los datos contenidos, densidades, segregaciones, agregaciones, sesgos mestadisticos descriptivos, estadisticos de dispersion, con respecto a variables y vector objetivo (income). 
* Describir la base de datos y re-definirla para adecuar sus valores, columnas, diseños, caracteres, etc... para no tener inconvenientes en el proceso.
    
* *En este documento se analizan las distintas variables que pueden afectar el    ingreso de cada individuo.*
* *Para ello primero se analiza la base de datos entregada, se cumplira con la recodificacion solicitada además de los datos perdidos, y con formatos que deban ser modificados para no tener problemas mas adelante.*
    - Para la variable income y todas las demas variables categoricas que se deberan recodificar en binarias se aplica la metodologia Binary Encoding, con el fin de generar vectores perpendiculares entre si, sin colinealidad, es decir se genera  una matriz fullrank.
* *Se utilizan visualizacion de variables, medidas de tendencia central, de dispersion, densidad, frecuencia, correlacion, graficos de barras, histogramas, boxplot, entre otros para definir el comportamiento de las variables.*
* *Luego mediante inferencia estadistica y modelacion logaritmica se analizará la correlacion entre variables y se seleccionaran las variables que continuan en analisis vs las que no.*
* *Se generan modelos (log) depurados por medio de las metricas principales (BIC, Rsqr, P-value) para estudiar el comportamiento probabilistico de manera general.*
* *Finalmente se crean y entrenan 3 modelos previamente definidos, para comparar sus metricas y evaluar su desempeño predictivo frente al vector objetivo 'income'. Con el fin de obtener el mejor modelo de clasificacion capaz de predecir si un individuo presenta salario por sobre o bajo los 50.000 usd anuales, en base a una serie de atributos sociodemograficos.*
#### Métricas
* Metricas Descriptivas: 
    - Indicadores BIC AUC: para medir la cantidad de parametros utilizados en los modelos. Mientras menor sea mejor.
    - P-Value: que nos indica si se puede rechazar la hipotesis nula. Ho indica que no se tienen valores significativos, por ende si se rechaza se tiene significancia en la variable o modelo de estudio.
    - Log-verosimilitud: Nos indica que tan bueno es el modelo en base a su verosimilitud, mientras mayor sea el indicador mejor.
    - R-squared: Capacidad explicativa del modelo y/o conjunto de variables.
* Metricas Prediccion:
    - Acuracy: basado en la matriz de confusion, mide los casos predichos correctamente vs el total de casos. como benchmark podemos considerar 1 dividido en la cantidad de categorias en el vector objetivo, es decir 1/2=0.5.
    - Precision: Proporcion de predicciones correctas entre los positivos.
    - Recall o sensibilidad: Fraccion de verdaderos positivos dentro del modelo. 
    - F1: Metrica que fusiona a Precision y Recall y penaliza cuando una de ellas es baja. se le dara mas énfasis.
    - Curva ROC: curva por sobre la bisectriz sera un buen modelo, ya que indica que predice mas verdaderos positivos que falsos positivos.
    - AUC: Area bajo la curva, nos indica el desempeño del modelo vs un mecanismo de clasificacion aleatoria.

### Cross Validation

```python
eval_metrics = ['f1', 'precision', 'recall', 'accuracy']
plt.subplots(1, len(eval_metrics))
plt.xticks(rotation=70)
for n, i in enumerate(eval_metrics):
    tmp_1 = cross_val_score(LogisticRegression(), df_mod.loc[:, var_mod2],
                    df_mod.loc[:,'income_may_50K'], cv = 10, scoring = i)
    tmp_2 = cross_val_score(LogisticRegression(), df_mod.loc[:, var_mod3],
                        df_mod.loc[:,'income_may_50K'], cv=10, scoring=i)
    tmp_3 = cross_val_score(LogisticRegression(), df_mod.loc[:, var_mod4], 
                        df_mod.loc[:,'income_may_50K'], cv=10, scoring=i)
    tmp = pd.DataFrame({'Reg 95%': tmp_1,
    '10 mejores': tmp_2,
    '4 best corr': tmp_3}).unstack().reset_index()
    tmp.rename(columns={'level_0':'order',
    'level_1': 'num',
    '0': 'score'}, inplace=True)
    plt.subplot(1, 4, n+1)
    sns.boxplot(tmp.iloc[:,0],
    tmp.iloc[:, 2])
    sns.despine()
    plt.xticks(rotation=45, ha='right')
    plt.xlabel(' ')
    plt.ylabel(' ')
    plt.title(i.capitalize())
```

* El modelo depurado de regresores muestra  ser por lejos el mejor en cuanto a prediccion, sin embargo al utilizar tantos regresores es un modelo poco parsimonioso y poco flexible.
* Entre el modelo de los 10 mejores y el modelo de las mejores correlaciones, los 10 mejores presentan una media mayor en todos los indicadores respecto a las iteraciones de prueba, menos en presicion donde pareciera estar mejor el modelo de las 4 best corr.
* En F1 muestra ser mucho mejor el modelo de los 10 mejores, por ende este seria el mejor modelo que equilibra la prediccion y la parsimonía, sin quitarle meritos al modelo de las mejores 4 correlaciones que para usar pocas variables predice en gran nivel.
